```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
```

```python
Federer = pd.read_csv('Roger_Federer.csv')
Nadal = pd.read_csv('Rafael_Nadal.csv')
Djokovic = pd.read_csv('Novak_Djokovic.csv')
```

```python
winpercent = pd.DataFrame({'Federer': Federer.Win.values, 
                   'Nadal': Nadal.Win.values,
                   'Djokovic': Djokovic.Win.values}, 
                  index=Federer.Tournament)
winpercent.plot()
```

```python
F=Federer.iloc[:,7:-1]
N=Nadal.iloc[:,3:-1]
D=Djokovic.iloc[:,1:-1]
```

```python
w=0
l=0
for i in F:
    for j in range(4):
        if F[i][j]=='W' and N[i][j]=='F':
            w += 1
        elif F[i][j]=='F' and N[i][j]=='W':
            l += 1
print('Percentage of Federer win Nadal in final round is {:.2f}'.format(w/(w+l)*100))
```

```python
w=0
l=0
for i in F:
    for j in range(4):
        if F[i][j]=='W' and D[i][j]=='F':
            w += 1
        elif F[i][j]=='F' and D[i][j]=='W':
            l += 1
print('Percentage of Federer win Djokovic in final round is {:.2f}'.format(w/(w+l)*100))
```
